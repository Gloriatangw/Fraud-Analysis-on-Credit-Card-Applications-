```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
 from scipy import stats
```

```python
data=pd.read_csv("train.csv")
```

```python
numberof1=sum(data.loc[:,"fraud"])
numberof0=data.iloc[:,0].size-numberof1
dfname=data.columns.values
result=[]
```

```python
for i in dfname:
    if(i !="fraud" ):
        temp0=data.loc[data.loc[:,"fraud"]==0,("fraud",i)]
        temp1=data.loc[data.loc[:,"fraud"]==1,("fraud",i)]
        #print(i)
        result0=pd.DataFrame({i:temp0.groupby(temp0[i]).size().index,"0":temp0.groupby(temp0[i]).size().values})
        result1=pd.DataFrame({i:temp1.groupby(temp1[i]).size().index,"1":temp1.groupby(temp1[i]).size().values})
        joined=pd.merge(result1,result0,how='outer')
        joined.fillna(0)
        joined.loc[:,"0"]=joined.loc[:,"0"]/numberof0
        joined.loc[:,"1"]=joined.loc[:,"1"]/numberof1
        numofrow=joined.iloc[:,0].size-1
        for index in range(0,numofrow):
            joined.loc[index+1,"0"]=joined.loc[index+1,"0"]+joined.loc[index,"0"]
            joined.loc[index+1,"1"]=joined.loc[index+1,"1"]+joined.loc[index,"1"]
        print(joined.loc[:,"1"])
        a=stats.ks_2samp(np.array(joined.loc[:,"0"]), np.array(joined.loc[:,"1"]))
        result.append([i,a.statistic])
        #print(stats.ks_2samp(np.array(joined.loc[:,"0"]), np.array(joined.loc[:,"1"].)))
```

```python
result=pd.DataFrame(result)
result.columns=["variables","D"]
result=result.sort_values('D', ascending=False)
result.to_csv("ksresult.csv")
result.iloc[0:40,:].to_csv("top 40 after ks.csv")
```
